standard_inchi_key를 기준으로 PubChem, STITCH 매핑하기

```python
import pandas as pd
import numpy as np
import time
import requests
from tqdm import tqdm
```

```python
terp = df[df["np_classifier_pathway"] == "Terpenoids"].copy()

print("Terpenoids:", len(terp))
print("InChIKey 있음:", terp["standard_inchi_key"].notna().sum())
print("name 있음:", terp["name"].notna().sum())
```

```python
# PubChem mapping 용 테이블 만들기
```

```python
terp_map = terp[
    [
        "identifier",
        "name",
        "iupac_name",
        "canonical_smiles",
        "standard_inchi",
        "standard_inchi_key",
        "np_classifier_pathway",
        "np_classifier_superclass",
        "np_classifier_class",
        "chemical_super_class",
        "chemical_class",
        "chemical_sub_class"
    ]
].copy()

terp_map = terp_map.dropna(subset=["standard_inchi_key"]).copy()

terp_map = terp_map.drop_duplicates(subset=["standard_inchi_key"]).copy()

print("Unique InChIKey:", len(terp_map))
terp_map.head()
```

```python
from tqdm import tqdm
import pandas as pd
import requests
import time
import os

def inchikey_to_pubchem_cid(inchikey):
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/inchikey/{inchikey}/cids/TXT"
    try:
        r = requests.get(url, timeout=10)
        if r.status_code == 200:
            cids = r.text.strip().splitlines()
            return cids[0] if len(cids) > 0 else None
        return None
    except Exception:
        return None
```

```python
unique_keys = terp_map["standard_inchi_key"].dropna().drop_duplicates().tolist()

save_path = "terpenoid_inchikey_to_pubchem_cid_partial.csv"

# 이전에 저장된 게 있으면 이어서 진행
if os.path.exists(save_path):
    done_df = pd.read_csv(save_path)
    done_keys = set(done_df["standard_inchi_key"])
    results = done_df.to_dict("records")
else:
    done_keys = set()
    results = []

remaining_keys = [k for k in unique_keys if k not in done_keys]

print("전체 InChIKey:", len(unique_keys))
print("이미 완료:", len(done_keys))
print("남은 개수:", len(remaining_keys))
```

```python
for i, key in enumerate(tqdm(remaining_keys)):
    cid = inchikey_to_pubchem_cid(key)
    results.append({
        "standard_inchi_key": key,
        "pubchem_compound_id": cid
    })

    # 500개마다 중간 저장
    if (i + 1) % 500 == 0:
        pd.DataFrame(results).to_csv(save_path, index=False)

    time.sleep(0.1)

# 최종 저장
pubchem_mapping = pd.DataFrame(results)
pubchem_mapping.to_csv("terpenoid_inchikey_to_pubchem_cid.csv", index=False)

print("완료:", len(pubchem_mapping))
print("CID 매핑 성공:", pubchem_mapping["pubchem_compound_id"].notna().sum()) 
```

```python
# 기존 COCONUT 정보와 합치기

terp_pubchem = terp_map.merge(
    pubchem_mapping,
    on="standard_inchi_key",
    how="left"
)

print("전체 Terpenoid:", len(terp_pubchem))
print("PubChem CID 매핑 성공:", terp_pubchem["pubchem_compound_id"].notna().sum())
print("PubChem CID 매핑 실패:", terp_pubchem["pubchem_compound_id"].isna().sum())
```

```python
# 3번 csv 생성

terpenoid_chemical_pubchem = terp_pubchem[
    terp_pubchem["pubchem_compound_id"].notna()
].copy()

terpenoid_chemical_pubchem["chemical"] = terpenoid_chemical_pubchem["name"]
terpenoid_chemical_pubchem["chemical"] = terpenoid_chemical_pubchem["chemical"].fillna(
    terpenoid_chemical_pubchem["iupac_name"]
)
terpenoid_chemical_pubchem["chemical"] = terpenoid_chemical_pubchem["chemical"].fillna(
    terpenoid_chemical_pubchem["identifier"]
)

terpenoid_chemical_pubchem = terpenoid_chemical_pubchem[
    ["chemical", "pubchem_compound_id"]
].drop_duplicates()

terpenoid_chemical_pubchem.to_csv(
    "terpenoid_chemical_pubchem_compound_id.csv",
    index=False
)

print("최종 CSV 행 수:", len(terpenoid_chemical_pubchem))
terpenoid_chemical_pubchem.head()
```
